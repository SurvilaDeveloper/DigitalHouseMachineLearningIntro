```python
df = pd.DataFrame(data=ld.data, columns=ld.feature_names)
df.head()
```

```python
df.rename(columns={
    'age': 'edad',
    'sex': 'sexo',
    'bmi': 'índice_de_masa_corporal',
    'bp': 'presión_arterial_media',
    's1': "colesterol_total",
    's2': "colesterol_LDL_malo",
    's3': "colesterol_HDL_bueno",
    's4': "colesteriol_total/colesterol_HDL",
    's5': "log_de_Triglicéridos",
    's6': "glucosa"}, inplace=True)
df.head(3)
```

```python
X = df
y = ld.target
```

```python
# Aquí quiero agregar visualización de todas las variables del dataset
# en función de la variable objetivo, que es la glucosa

sns.set_palette("pastel")
plt.figure(figsize=(9, 9))
plt.subplot(3, 3, 1)
sns.scatterplot(data=df, x='edad', y=y)
plt.title('Edad vs Diabetes')
plt.subplot(3, 3, 2)
sns.scatterplot(data=df, x='sexo', y=y)
plt.title('Sexo vs Diabetes')
plt.subplot(3, 3, 3)
sns.scatterplot(data=df, x='índice_de_masa_corporal', y=y)
plt.title('Índice de Masa Corporal vs Diabetes')
plt.subplot(3, 3, 4)
sns.scatterplot(data=df, x='presión_arterial_media', y=y)
plt.title('Presión Arterial Media vs Diabetes')
plt.subplot(3, 3, 5)
sns.scatterplot(data=df, x='colesterol_total', y=y)
plt.title('Colesterol Total vs Diabetes')
plt.subplot(3, 3, 6)
sns.scatterplot(data=df, x='colesterol_LDL_malo', y=y)
plt.title('Colesterol LDL Malo vs Diabetes')
plt.subplot(3, 3, 7)
sns.scatterplot(data=df, x='colesteriol_total/colesterol_HDL', y=y)
plt.title('colesteriol_total/colesterol_HDL vs Diabetes')
plt.subplot(3, 3, 8)
sns.scatterplot(data=df, x='log_de_Triglicéridos', y=y)
plt.title('log_de_Triglicéridos vs Diabetes')
plt.subplot(3, 3, 9)
sns.scatterplot(data=df, x='glucosa', y=y)
plt.title('Glucosa vs Diabetes')
plt.tight_layout()
plt.show()
```

```python
datacompleto = pd.DataFrame(data=X)
datacompleto['diabetes'] = y

plt.figure(figsize=(6,6))
sns.heatmap(datacompleto.corr()[['diabetes']], annot=True)
plt.show()
plt.figure(figsize=(6,6))
sns.heatmap(X.corr(), annot=True)
plt.show()

```

```python
correlacion_con_target = datacompleto.corr()['diabetes'].sort_values(key=abs, ascending=False)
print(correlacion_con_target)
```

```python
umbral = 0.3
features_seleccionados = abs(correlacion_con_target[abs(correlacion_con_target) > umbral]).index.tolist()
print(features_seleccionados)
```

```python
df_filtrado = datacompleto[features_seleccionados]
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(df_filtrado.corr(), annot=True)
plt.show()
```

```python
X = df_filtrado.drop('diabetes', axis=1, inplace=False)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Error cuadrático medio: {mse:.2f}")
print(f"R^2: {r2:.2f}")
```

```python
# Visualización de los resultados
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r-')
plt.xlabel('Valores reales')
plt.ylabel('Predicciones')
plt.title('Predicciones vs Valores reales')
plt.show()
```
